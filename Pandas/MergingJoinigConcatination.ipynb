```python
import numpy as np
import pandas as pd
from pandas.core.interchange.dataframe_protocol import DataFrame
```

Merging Two DataFrames

```python
empoloyees = {
    'employee_id' : [1,2,3,4,5],
    'name' : ['Navneet','Rohan','Rahul','Raj','Abhishek'],
    'department'  : ['HR','IT','CS','Finance','HR']
}
df1 = pd.DataFrame(empoloyees)
df1
```

```python
salaries = {
    'employee_id' : [1,2,3,6,7],
    'salary' : [30000,25000,20000,15000,10000],
    'bonus'  : [5000,7000,9000,11000,14000]
}
df2 = pd.DataFrame(salaries)
df2
```

```python
pd.merge(df1,df2,on ='employee_id', how='inner')
```

```python
pd.merge(df1,df2,on ='employee_id', how='outer')
```

```python
pd.merge(df1,df2,on ='employee_id', how='left')
```

```python
pd.merge(df1,df2,on ='employee_id', how='right')
```

Concatination Two DataFrames

```python
df1 = pd.DataFrame({
    'A': ['A0', 'A1', 'A2'],
    'B': ['B0', 'B1', 'B2'],
    'C': ['C0', 'C1', 'C2']
})

df2 = pd.DataFrame({
    'A': ['A3', 'A4', 'A5'],
    'B': ['B3', 'B4', 'B5'],
    'C': ['C3', 'C4', 'C5']
})
```

```python
df1
```

```python
df2
```

```python
pd.concat([df1,df2])
```

```python
pd.concat([df1,df2], axis=1)
```

Joining Of Two DataFrames

```python
df1 = pd.DataFrame({
    'name': ['Alice', 'Bob', 'Charlie']
}, index=[1, 2, 3])

# Second DataFrame
df2 = pd.DataFrame({
    'score': [85, 90, 75]
}, index=[2, 3, 4])
```

```python
df1
```

```python
df2
```

```python
df1.join(df2, how='outer')
```
