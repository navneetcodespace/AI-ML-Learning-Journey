### **1. Basics of Probability**

#### **1(a) Coin Toss Simulation**

```python
import random

# Number of trials
trials = 10000

heads = 0
tails = 0

for _ in range(trials):
    toss = random.choice(["H", "T"])

    if toss == "H":
        heads += 1
    else:
        tails += 1

# Experimental probabilities
prob_heads = heads / trials
prob_tails = tails / trials

print("Heads:", heads)
print("Tails:", tails)
print("Probability of Heads:", prob_heads)
print("Probability of Tails:", prob_tails)
```

**Explanation

This program simulates tossing a fair coin 10,000 times.
The experimental probability is calculated by dividing the number of successful outcomes by the total number of trials.
As the number of trials increases, the probabilities approach the theoretical values of 0.5 for heads and tails.**

1(b) Rolling Two Dice

```python
import random

trials = 10000
sum_7_count = 0

for _ in range(trials):
    die1 = random.randint(1, 6)
    die2 = random.randint(1, 6)

    if die1 + die2 == 7:
        sum_7_count += 1

probability = sum_7_count / trials

print("Count of sum 7:", sum_7_count)
print("Experimental Probability:", probability)
```

Explanation

This program simulates rolling two dice 10,000 times and checks how often the sum is equal to 7.
The probability is estimated using simulation and should be close to the theoretical probability of 6/36 = 0.1667.

2. Probability of Getting At Least One “6”

```python
import random
def probability_at_least_one_six(trials=10000):
    success = 0

    for _ in range(trials):
        found_six = False

        for _ in range(10):
            roll = random.randint(1, 6)

            if roll == 6:
                found_six = True
                break

        if found_six:
            success += 1

    return success / trials

result = probability_at_least_one_six()

print("Estimated Probability:", result)
```

Explanation

This program estimates the probability of getting at least one 6 when a die is rolled 10 times.
A trial is considered successful if at least one roll gives a 6.
The experimental probability is obtained by dividing successful trials by total trials.

3. Conditional Probability and Bayes’ Theorem

```python
import random

balls = ["Red"] * 5 + ["Green"] * 7 + ["Blue"] * 8

trials = 1000

previous_ball = None

blue_then_red = 0
previous_blue = 0

red_count = 0
blue_count = 0

for _ in range(trials):
    current_ball = random.choice(balls)

    if current_ball == "Red":
        red_count += 1

    if current_ball == "Blue":
        blue_count += 1

    # Conditional probability
    if previous_ball == "Blue":
        previous_blue += 1

        if current_ball == "Red":
            blue_then_red += 1

    previous_ball = current_ball

# P(Red | Previous Blue)
conditional_probability = blue_then_red / previous_blue

# Bayes theorem verification
P_red = red_count / trials
P_blue = blue_count / trials

print("P(Red | Previous Blue):", conditional_probability)
print("P(Red):", P_red)
print("P(Blue):", P_blue)
```

Explanation

This simulation repeatedly draws balls with replacement from a bag containing red, green, and blue balls.
The program estimates the probability of drawing a red ball given that the previous ball was blue.

Since the draws are independent (because the ball is replaced), the conditional probability should be close to the normal probability of drawing a red ball.

4. Discrete Random Variable

```python
import numpy as np

values = [1, 2, 3]
probabilities = [0.25, 0.35, 0.40]

sample = np.random.choice(values, size=1000, p=probabilities)

mean = np.mean(sample)
variance = np.var(sample)
std_dev = np.std(sample)

print("Empirical Mean:", mean)
print("Variance:", variance)
print("Standard Deviation:", std_dev)
```

Explanation

This program generates 1000 random samples from a discrete probability distribution using NumPy.
The empirical mean, variance, and standard deviation are calculated to summarize the behavior of the random variable.

5. Continuous Random Variables – Exponential Distribution

```python
import numpy as np
import matplotlib.pyplot as plt

# Mean = 5
scale = 5

samples = np.random.exponential(scale=scale, size=2000)

# Histogram
plt.hist(samples, bins=30, density=True)

# PDF Overlay
x = np.linspace(0, max(samples), 1000)

pdf = (1/scale) * np.exp(-x/scale)

plt.plot(x, pdf)

plt.title("Exponential Distribution")
plt.xlabel("Value")
plt.ylabel("Density")

plt.show()
```

Explanation

This program generates 2000 samples from an exponential distribution with mean 5.
A histogram is used to visualize the sampled data, and the theoretical probability density function (PDF) is plotted on top for comparison.

6. Central Limit Theorem (CLT)

```python
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Generate uniform data
uniform_data = np.random.uniform(0, 1, 10000)

# Step 2: Draw samples and calculate means
sample_means = []

for _ in range(1000):
    sample = np.random.choice(uniform_data, size=30)
    sample_means.append(np.mean(sample))

# Plot Uniform Distribution
plt.hist(uniform_data, bins=30, density=True)
plt.title("Uniform Distribution")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()

# Plot Distribution of Sample Means
plt.hist(sample_means, bins=30, density=True)
plt.title("Distribution of Sample Means")
plt.xlabel("Sample Mean")
plt.ylabel("Density")
plt.show()
```

Explanation

This program demonstrates the Central Limit Theorem.
Although the original data follows a uniform distribution, the distribution of sample means becomes approximately normal when many samples are taken.
This shows that sample means tend toward a normal distribution regardless of the original distribution shape.
