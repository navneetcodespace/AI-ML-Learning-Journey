### **Assignment 1 — NumPy**

```python
# NumPy Assignment
# This notebook demonstrates NumPy array creation, manipulation, mathematical operations, reshaping, and slicing.
```

```python
import numpy as np
```

#### **Question 1 — 1D Array Operations**

```python
## 1D Array Operations
# In this section, we create a one-dimensional NumPy array and perform statistical and conditional operations.
```

```python
# Create array from 1 to 20
arr1 = np.arange(1, 21)

print("Array:")
print(arr1)

# Statistical operations
print("\nSum:", np.sum(arr1))
print("Mean:", np.mean(arr1))
print("Median:", np.median(arr1))
print("Standard Deviation:", np.std(arr1))

# Indices of elements greater than 10
indices = np.where(arr1 > 10)

print("\nIndices of elements greater than 10:")
print(indices)
```

### **Question 2 — 2D Array Operations**

```python
## 2D Array Operations
# This section demonstrates matrix creation, transpose operations, and row-wise and column-wise calculations.
```

```python
# Create 4x4 array
arr2 = np.arange(1, 17).reshape(4, 4)

print("4x4 Array:")
print(arr2)

# Transpose
print("\nTranspose:")
print(arr2.T)

# Row-wise sum
print("\nRow-wise Sum:")
print(np.sum(arr2, axis=1))

# Column-wise sum
print("\nColumn-wise Sum:")
print(np.sum(arr2, axis=0))
```

#### **Question 3 — Random Arrays**

```python
## Random Array Operations
# Here, we generate random matrices and perform arithmetic and matrix multiplication operations.
```

```python
# Random 3x3 arrays
arr3 = np.random.randint(1, 21, (3, 3))
arr4 = np.random.randint(1, 21, (3, 3))

print("Array 1:")
print(arr3)

print("\nArray 2:")
print(arr4)

# Element-wise operations
print("\nAddition:")
print(arr3 + arr4)

print("\nSubtraction:")
print(arr3 - arr4)

print("\nMultiplication:")
print(arr3 * arr4)

# Dot product
print("\nDot Product:")
print(np.dot(arr3, arr4))
```

#### **Question 4 — Reshaping and Slicing**

```python
## Reshaping and Slicing
# This section demonstrates reshaping a one-dimensional array into a matrix and extracting specific rows and columns.
```

```python
# Create 1D array
arr5 = np.arange(1, 13)

# Reshape into 3x4
reshaped = arr5.reshape(3, 4)

print("Reshaped Array:")
print(reshaped)

# Slice first two rows and last two columns
sliced = reshaped[:2, -2:]

print("\nSliced Array:")
print(sliced)
```

### **Assignment 2 — Pandas**

```python
import pandas as pd
```

#### **Create DataFrame**

```python
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'Diana', 'Eve'],
    'Age': [24, 27, 22, 32, 29],
    'Department': ['HR', 'Finance', 'IT', 'Marketing', 'HR'],
    'Salary': [45000, 54000, 50000, 62000, 47000]
}

df = pd.DataFrame(data)

print(df.head())
```

#### **Summary Statistics**

```python
print(df[['Age', 'Salary']].describe())
```

#### **Average Salary in HR**

```python
hr_avg_salary = df[df['Department'] == 'HR']['Salary'].mean()

print("Average HR Salary:", hr_avg_salary)
```

#### **Add Bonus Column**

```python
df['Bonus'] = df['Salary'] * 0.10

print(df)
```

#### **Filter Employees Between 25 and 30**

```python
filtered_df = df[(df['Age'] >= 25) & (df['Age'] <= 30)]

print(filtered_df)
```

#### **Group by Department**

```python
dept_avg_salary = df.groupby('Department')['Salary'].mean()

print(dept_avg_salary)
```

#### **Sort by Salary and Save CSV**

```python
sorted_df = df.sort_values(by='Salary')

print(sorted_df)

sorted_df.to_csv('sorted_salary.csv', index=False)
```

### **Assignment 3 — Matplotlib**

```python
import matplotlib.pyplot as plt
import numpy as np
```

#### **Line Plot**

```python
x = [1, 2, 3, 4, 5]
y = [10, 15, 25, 30, 50]

plt.plot(x, y, marker='o')

plt.title("Line Plot Example")
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.grid(True)

plt.show()
```

#### **Bar Graph**

```python
students = ['John', 'Jane', 'Alice', 'Bob']
marks = [75, 85, 60, 90]

plt.bar(students, marks, color=['blue', 'green', 'orange', 'red'])

plt.title("Student Marks")

plt.xlabel("Students")
plt.ylabel("Marks")

plt.show()
```

#### **Pie Chart**

```python
regions = ['North America', 'Europe', 'Asia', 'Others']
revenue = [45, 25, 20, 10]

explode = [0.1, 0, 0, 0]

plt.pie(
    revenue,
    labels=regions,
    explode=explode,
    autopct='%1.1f%%',
    shadow=True
)

plt.title("Revenue Distribution by Region")

plt.show()
```

#### **Histogram**

```python
# Generate random integers
data = np.random.randint(1, 101, 1000)

plt.hist(data, bins=10)

plt.title("Histogram of Random Numbers")
plt.xlabel("Value Range")
plt.ylabel("Frequency")

plt.show()
```
