```python
a = np.array([1,2,3,4,5])
b = np.array([6,7,8,9,10])
```

```python
#There are three type of stack
# 1 . Vertical Stack
np.vstack((a,b))

# 2. Horizontal Stack
np.hstack((a,b))

# 3. Column stack
np.column_stack((a,b))
```

Splitting array

```python
c = np.arange(1,17).reshape(4,4)
c
```

```python
# Horizontal Split
np.hsplit(c,2)

# Vertical Split
d = np.vsplit(c,4)

for i in d:
    print(i)
```
