**Case Study - Insurance Claim Records using numpy**

Consider an insurance comapany analyzing monthly claims made by policyholders:

```python
import numpy as np
```

```python
# Monthly claim amounts (IN INR) for diffrent policyholders
claims = np.array([
    [12000,15000,18000], #policyholder 1
    [10000,13000,17000], #policyholder 2
    [9000,11000,16000]   #policyholder 3
])

print("Insurance Claim Data")
print(claims)
```

```python
# Accessing element in NDArray
# Access claim amount of policyholder 3 for the first month

print("Claim amount of policyholder 3 for 1 month : ", claims[2,0])
print("Claim amount of policyholder 2 for 3 month : ", claims[1,2])
```

**Array Operation in Numpy**

```python
# Total claim amount per policyholder

total_claims = np.sum(claims, axis=1) # 1 for row wise , 0 for column wise
print("Total claim per policy holder : ", total_claims)
```

```python
# Average claim amount per months

avg_claim = np.average(claims, axis=0)
print("Average claim amount per months are : ", avg_claim)
```

```python
# Find claims greater than 15000

high_claims = claims[claims > 15000]
print("High vlue Claims : ",high_claims)
```
