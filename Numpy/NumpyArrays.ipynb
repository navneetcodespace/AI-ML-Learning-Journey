```python
arr1 = np.random.randint(6) #it will generate the 6 random integer
arr2 = np.random.randint(6,10) #it will generate the random integer between 6,10
arr3 = np.random.randint(10,30,10) #it will generate the array of 2 random integer between 6,10
print(arr1)
print(arr2)
print(arr3)
```
