```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Matplotlib Basics

```python
x = np.linspace(0,5,11)
x
```

```python
y = x**2
y
```

```python
plt.plot(x,y)
plt.title("Hello")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
```

**Subplot in Matplotlib** -> Subplot ka mtlb hai ek sath multiple plots banana

```python
plt.subplot(2,2,1)
plt.plot(y,x)
plt.subplot(2,2,2)
plt.plot(x**2, x**2)
plt.subplot(2,2,3)
plt.plot(y**2, y**2)
plt.subplot(2,2,4)
plt.plot(x,y)
```

Object Oriented Way

```python
fig = plt.figure( )
axis_1 = fig.add_axes([0.1,0.1,0.8,0.8])
axis_1.plot(x,y)

axis_2 = fig.add_axes([0.5,0.5,0.6,0.6])
axis_2.plot(x,y)
```

Types of plot


```python
plt.scatter(x,y)
```

```python
plt.hist(y)
```

```python
plt.boxplot(x)
```

Saving the plot

```python
fig = plt.figure( )
axis_1 = fig.add_axes([0.1,0.1,0.8,0.8])
axis_1.plot(x,y)

plt.savefig("Sample.png")
```
