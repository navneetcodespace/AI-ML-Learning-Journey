```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sms
```

Distribution Plot

```python
df = sms.load_dataset('tips')
df
```

```python
df['size'].unique()
```

```python
#displot / histplot
sms.displot(df['total_bill'],bins = 30)
```

```python
# kernal density estimater
```

```python
plt.subplot(1,2,1)
sms.histplot(df['total_bill'],bins = 20, kde = True)
plt.subplot(1,2,2)
sms.histplot(df['tip'],bins = 20, kde = True)
```

```python
# Joint Plot

sms.jointplot(x = 'total_bill', y = 'tip', data = df, kind = 'scatter')
```
